```python
import pymongo
import pandas as pd
```

```python
client = pymongo.MongoClient('')
db = client.NOME # nome del database
```

```python
doc = db.impianti.find_one()
doc
```

```python
df1 = pd.DataFrame(list(db.impianti.find()))
df1
```

1. Visualizzare l'elenco degli impianti accesi. Visualizzare solo il nome e l'indirizzo

```python
result = db.impianti.find({"Sit_StatoImpianto": "Acceso"}, {"_id": 0, "Sit_NomeImpianto": 1, "INDIRIZZO": 1})
df = pd.DataFrame(list(result))
df
```

2. Visualizzare gli indirizzi in cui si trovano gli impianti dell'ATM

```python
result = db.impianti.find({"gestore": {"$regex": "A.T.M."}}, {"INDIRIZZO": 1, "_id": 0})
df = pd.DataFrame(list(result))
df
```

3. Visualizzare l'elenco degli impianti della Vodafone e della TIM

```python
result = db.impianti.find({'$or': [{'gestore': "VODAFONE"}, {"gestore": {"$regex": "TIM"}}]})
df = pd.DataFrame(list(result))
df
```

4. Visualizzare gli indirizzi in cui c'è più di un impianto

```python
result = db.impianti.aggregate([{"$group": {"_id": "$INDIRIZZO", "total": {"$sum": 1}}},
                                {"$match": {"total": { "$gt": 1}}},
                                {"$project": {"_id": 1, "total": 0}}])
df2 = pd.DataFrame(list(result))
df2
```

5. Visualizzare i gestori che sono società per azioni. Ordinare in ordine alfabetico



```python
result = db.impianti.find({'gestore': {"$regex": "S.p.A."}}, {"gestore": 1, "_id": 0}).sort("gestore", 1)
df = pd.DataFrame(list(result))
df
```

6. Visualizzare il numero di impianti per ogni gestore

```python
result = db.impianti.aggregate([{"$group": {"_id": "$gestore", "total": {"$sum": 1}}}])
df2 = pd.DataFrame(list(result))
df2
```

7. Visualizzare i gestori che hanno più di 100 impianti. Ordinare in ordine decrescente a partire dal gestore che ha più impianti

```python
result = db.impianti.aggregate([{"$group": {"_id": "$gestore", "total": {"$sum": 1}}},
                                {"$match": { "total": {"$gt": 100}}},
                                {"$sort": {"total": -1 }}])
df2 = pd.DataFrame(list(result))
df2
```

8. Visualizzare i gestori che hanno impianti all'aeroporto di Linate

```python
result = db.impianti.find({"INDIRIZZO": {"$regex": "LINATE"}}, {"_id": 0, "gestore": 1, "INDIRIZZO": 1})
df = pd.DataFrame(list(result))
df
```

9. Visualizzare i gestori di cui non si conosce il tipo della via in cui si trovano i loro impianti

```python
result = db.impianti.find({"Tv_TipoVia": {"$regex": "---"}})
df = pd.DataFrame(list(result))
df
```

10. Visualizzare gli impianti che si trovano a Cimiano

```python
result = db.impianti.find({"NIL": {"$regex": "CIMIANO"}}, {"_id": 0, "Sit_NomeImpianto": 1, "NIL": 1})
df = pd.DataFrame(list(result))
df
```

11. Visualizzare gli impianti che non sono accesi e chi si trovano a Cimiano e nel quartiere Adriano

```python
result = db.impianti.find({'$and': [{'Sit_StatoImpianto': {'$not': {'$eq': 'Acceso'}}}, {'$or': [{'NIL': {'$regex': 'CIMIANO'}}, {'NIL': {'$regex': 'Adriano'}}]}]})
df11 = pd.DataFrame(list(result))
df11[['Sit_NomeImpianto', 'Sit_StatoImpianto', 'NIL']]
```

12. Visualizzare il quartiere con più impianti

```python
result = db.impianti.aggregate([{"$group": {"_id": "$MUNICIPIO", "total": {"$sum": 1}}},
                                {"$sort": {"_id": -1}}])
df2 = pd.DataFrame(list(result))
df2
```
