```python
import pymongo
import pandas as pd
```

```python
client = pymongo.MongoClient('')
db = client.NOME # nome del database
```

# **INSERT**

```python
import datetime
post = {"author": "Mike",
        "text": "My first blog post!",
        "tags": ["mongodb", "python", "pymongo"],
        "date": datetime.datetime.utcnow()}
```

```python
post_id = db.posts_test.insert_one(post).inserted_id
post_id
```

```python
df1 = pd.DataFrame(list(db.posts_test.find()))
df1
```

PIU POST (PIU DI UNO)

```python
new_posts = [{"author": "Mike",
              "text": "Another post!",
              "tags": ["bulk", "insert"],
              "date": datetime.datetime(2009, 11, 12, 11, 14)},
             {"author": "Eliot",
              "title": "MongoDB is fun",
              "text": "and pretty easy too!",
              "date": datetime.datetime(2009, 11, 10, 10, 45)}]
```

```python
post_id = db.posts_test.insert_many(new_posts).inserted_ids
post_id
```

```python
df1 = pd.DataFrame(list(db.posts_test.find()))
df1
```

```python
from bson.objectid import ObjectId
df1 = pd.DataFrame(list(db.posts_test.find({'_id': ObjectId('6245910193197e655390b7b1')})))
df1
```

#**UPDATE**

```python
result = db.posts_test.update_one({'_id': ObjectId('6245910193197e655390b7b1')}, {"$set": {"text": 'Post changed!!?'}})
result.modified_count
```

```python
df1 = pd.DataFrame(list(db.posts_test.find()))
df1
```

```python
result = db.posts_test.update_many({'author': "Mike"}, {"$set": {"author": 'Michael'}})
result.modified_count
```

```python
df1 = pd.DataFrame(list(db.posts_test.find()))
df1
```

```python
result = db.posts_test.update_many({'_id': ObjectId('6245910193197e655390b7b1')}, {"$push": {"tags": 'append'}})
result.modified_count
```

```python
df1 = pd.DataFrame(list(db.posts_test.find()))
df1
```

```python
result = db.posts_test.update_many({'_id': ObjectId('6245910193197e655390b7b1')}, {"$pull": {"tags": 'append'}})
result.modified_count
```

```python
df1 = pd.DataFrame(list(db.posts_test.find()))
df1
```

# **DELETE**

```python
result = db.posts_test.delete_one({"author": 'Michael'})
result.deleted_count
```

```python
df1 = pd.DataFrame(list(db.posts_test.find()))
df1
```

```python
result = db.posts_test.delete_many({"author": 'Eliot'}) 
result.deleted_count
```

```python
df1 = pd.DataFrame(list(db.posts_test.find()))
df1
```
