```
# estrae i documenti che hanno nella colonna "cuisine" il valore "Bakery" e nella colonna "borough" il valore "Brooklyn"

result = db.restaurants.find({'$and': [{"cuisine": "Bakery"}, {"borough": "Brooklyn"}]})
df2 = pd.DataFrame(list(result))
df2
```

```
# estrae i documenti che hanno nella colonna "cuisine" il valore "Bakery" o nella colonna "borough" il valore "Italian"

result = db.restaurants.find({'$or': [{"cuisine": "Bakery"}, {"cuisine": "Italian"}]})
df2 = pd.DataFrame(list(result))
df2
```

```
# estrae i documenti in cui è presente il valore "Bakery" e "Italian" nella colonna "cuisine"

result = db.restaurants.find({"cuisine": { '$in': ["Bakery", "Italian"]}})
df2 = pd.DataFrame(list(result))
df2
```

```
# estrae i documenti in cui nella colonna "borough" è presente il valore "Brooklyin" e nella colonna "cuisine" sono presenti i valori "Bakery" o "Italian"

result = db.restaurants.find({"borough": "Brooklyn", "cuisine": { '$in': ["Bakery", "Italian"]}})
df2 = pd.DataFrame(list(result))
df2
```

```
# estrae i documenti in cui nel campo "zipcode" che si trova nella colonna "address", è presenete il valore "10462"

result = db.restaurants.find({"address.zipcode": "10462"})
df2 = pd.DataFrame(list(result))
df2.head()
```

```
# estrae i documenti in cui nel campo "grade" che si trova nella colonna "grades", è presenete almeno un valore "C" e visualizza le colonne "grades" e "name"

result = db.restaurants.find({"grades.grade": "C"}, {"name": 1, "grades":1 })
df2 = pd.DataFrame(list(result))
df2.head()
```

#Funzioni di aggregazione, group by e having

```
# conta quanti documenti ci sono nella collection restaurant

result = db.restaurants.aggregate([
                                   {"$group": {"_id": "null", "count": {"$sum": 1}}}
                                   ])
df2 = pd.DataFrame(list(result))
df2.head()
```

```
# conta i ristoranti di ogni quartiere

result = db.restaurants.aggregate([
                                   {"$group": {"_id": "$borough", "total": {"$sum": 1}}}
                                   ])
df2 = pd.DataFrame(list(result))
df2.head()
```

```
# conta i ristoranti di ogni quartiere e li ordina in ordine decrescente

result = db.restaurants.aggregate([ 
                                   {"$group": {"_id": "$borough", "total": {"$sum": 1}}}, 
                                   {"$sort": { "total": -1 }} 
                                   ])
df2 = pd.DataFrame(list(result))
df2.head()
```

```
# estrae i quartieri con un numero di ristoranti maggiore di 3000 e ordina la lista in ordine decrescente

result = db.restaurants.aggregate([ 
                                   {"$group" : {"_id": "$borough", "total": {"$sum": 1}}}, 
                                   {"$match": {"total": {"$gt": 3000 }}}, 
                                   {"$sort": {"total": -1}} 
                                   ])
df2 = pd.DataFrame(list(result))
df2.head()
```

```
# estrae i quartieri con un numero di ristoranti maggiore di 70 e con cucina italiana e ordina la lista in ordine decrescente

result = db.restaurants.aggregate([ 
                                   {"$match": {"cuisine": "Italian"}}, 
                                   {"$group": {"_id": "$borough", "total": {"$sum": 1}}}, 
                                   {"$match": { "total": {"$gt": 70}}}, 
                                   {"$sort": { "total": -1}}])
df2 = pd.DataFrame(list(result))
df2.head()
```
