```
# gli appartamenti che costano più di 100 $ a notte
```

```
result = db.listingsAndReviews.find({"property_type": "Apartment", "price": {'$gt' : 100}},{"name":1,"property_type":1,"price":1})
df2 = pd.DataFrame(list(result))
df2
```

```
# gli appartamenti che costano tra i 100 $ e i 200 $ a notte
```

```
result = db.listingsAndReviews.find({"property_type": "Apartment", "price": {'$gt' : 100, '$lt' : 200 }},{"name":1,"property_type":1,"price":1})
df2 = pd.DataFrame(list(result))
df2
```

```
# gli appartamenti che hanno almeno 5 posti letto e che costano tra i 50 $ e 100 $ a notte
```

```
result = db.listingsAndReviews.find({"property_type": "Apartment", "beds" :{'$gt' : 4}, "price": {'$gt' : 50, '$lt' : 100 }}, {"name":1, "beds":1, "price":1,"property_type":1, "_id":0})
df2 = pd.DataFrame(list(result))
df2
```

```
# gli appartamenti in Portogallo
```

```
result = db.listingsAndReviews.find({"property_type": "Apartment",  "address.country": "Portugal"},{"name":1,"property_type":1, "address":1})
df2 = pd.DataFrame(list(result))
df2
```

```
# gli appartamenti che non si trovano in Portogallo e in Spagna
result = db.listingsAndReviews.find({"property_type": "Apartment",  "address.country": { '$not' : {'$in': ["Portugal", "Spain"]}}}, {"name":1,"property_type":1, "address":1})
df2 = pd.DataFrame(list(result))
df2
```

```
# gli appartamenti negli Stati Uniti che hanno il wifi
# ordinare in ordine decrescente sul prezzo per notte e prendere solo i primi 10 appartamenti
result = db.listingsAndReviews.find({"property_type": "Apartment",  '$and': [{"address.country": "United States"}, {"amenities" : "Wifi"}]}, {"name":1,"property_type":1, "address":1, "amenities":1, "price":1, "_id":0}).sort([("price", -1)])
df2 = pd.DataFrame(list(result))
df2.head(10)
```

```
# il numero di appartamenti in Portogallo
result = db.listingsAndReviews.aggregate([{ "$match": {"property_type": "Apartment"} },
                                          { "$match": {"address.country": "Portugal"} },
                                          { "$group" : {"_id": "$address.country", "total" : {"$sum" : 1}}}])
df2 = pd.DataFrame(list(result))
df2
```

```
# il numero di appartamenti nelle varie nazioni
result = db.listingsAndReviews.aggregate([{ "$match": {"property_type": "Apartment"} },
                                          { "$group" : {"_id": "$address.country", "total" : {"$sum" : 1}}}])
df2 = pd.DataFrame(list(result))
df2
```

```
# le nazioni con più di 600 appartamenti
result = db.listingsAndReviews.aggregate([{ "$match": {"property_type": "Apartment"} },
                                          { "$group" : {"_id": "$address.country", "total" : {"$sum" : 1}}},
                                          { "$match": { "total": { "$gt": 500 } } }])
df2 = pd.DataFrame(list(result))
df2
```

```
# gli appartamenti che hanno la parola 'beach' nella descrizione (google it...)
result = db.listingsAndReviews.aggregate([
                                          { "$match": { "description": {"$regex" : "beach"}}}])
df11 = pd.DataFrame(list(result))
df11[["description"]]
```
