```python
import pandas as pd
```

```python
df = pd.read_excel('https://github.com/wtitze/3E/blob/main/Medals.xlsx?raw=true')
df
```

Eliminare le colonne "Rank", "Total" e "Rank by Total".

```python
df.drop(columns = ["Rank", "Total", "Rank by Total"], inplace = True)
df
```

```python
df.rename(columns = {'Team/NOC': 'Team'}, inplace = True)
df
```

Creare una colonna con la somma delle medaglie di ogni paese.

```python
df['Total'] = df['Gold'] + df['Silver'] + df['Bronze']
df
```

Creare una colonna assegnando il valore 0.

```python
df['Colonna'] = 0
df
```

Visualizzare le nazioni che hanno più di 80 medaglie in totale.

```python
df[df['Total'] > 80][['Team']]
```

Visualizzare il numero medio di medaglie vinte dalle nazioni.

```python
df['Total'].mean()
```

Visualizzare le nazioni che hanno vinto più medaglie della media.

```python
df[df['Total'] > df['Total'].mean()]
```
