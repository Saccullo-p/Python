```python
movies[movies.Film.str.startswith("W")]
```

```python
movies[movies.Genre == 'Drama']['Audience score %'].mean()
```

11.visualizzare il numero di film per ogni genere


```python
movies.groupby(movies['Genre']).count()[['Film']]
```

```python
movies["Audience score %"] = movies["Audience score %"] * 0.9
movies
```
